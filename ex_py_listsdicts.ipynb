```python
# Create a shopping list with at least 5 items
groceries = ["kefir", "yogurt", "kraut", "kimchi", "kombucha"]
```

```python
# Print out the third item in the list
print(groceries[2])
```

```python
# Replace the first item with a different item
groceries[0] = "juice"
print(f"New Juice: {groceries}")
```

```python
# Add an item to the end of the list.append
groceries.append("non-juice")
print(groceries)
```

```python
# Create a dictionary of 3 grocery items with their prices
grocery_prices = {
    "yogurt": 7.89,
    "kraut": 10.59,
    "kimchi": 7.59,
}
```

```python
# Print a message showing one item and its price
print(f"The price of yogurt is ${grocery_prices["yogurt"]}")
```

```python
print(f"The price of {"yogurt"} is ${grocery_prices["yogurt"]}")
```

```python
# Add a new item to the dictionary with its price
grocery_prices.append("kombucha": 5.89)
```

```python
grocery_prices["kombucha"] = 5.89
print(grocery_prices)
```

```python
# Calculate and print the total cost of all items
total_cost = sum(grocery_prices)
```

```python
total_cost = sum(grocery_prices.values()) # the key was extracting the values
print(f"The total cost is ${total_cost:.2f}")
```

```python
total_cost = sum(float(price) for price in grocery_prices.values()) # .values extracts only values(prices)
print(f"The total cost is: ${total_cost:.2f}")                      # float(price) converts price strings into floats(not necessary here?)
```
